```python
import torch
from transformers import Wav2Vec2Model
```

```python
# Loaidng the pretrained model
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-xls-r-2b")
```

Base model is updated uning this checkpoint

```python
checkpoint_path = "/scratch/IITB/ai-at-ieor/23m1508/23m1508_backup/checkpoint_points/weights_535000_545000/multilingual_asr_model_535000_545000.pt"
```

```python
#######################################
# Use only one time since in this i have extatct the first 48 layers weights


checkpoint = torch.load(checkpoint_path, map_location = "cpu", weights_only = False)
    
state_dict = checkpoint['model_state_dict']  # 
```

```python
checkpoint
```

```python
base_model_state_dict = {
k.replace("projector.wav2vec2.", ""): v
for k, v in state_dict.items()
if k.startswith("projector.wav2vec2.")
}

# print('The keys of the base model state dict are :', base_model_state_dict.keys())
```

```python


missing, unexpected = model.load_state_dict(base_model_state_dict, strict = False)

print("Missing keys (likely your new CTC heads or projector):", missing)
print("Unexpected keys (should be empty):", unexpected)
```
